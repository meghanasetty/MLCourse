 # **REGRESSION DATA ANALYSIS**
 ### Introduction of Data
 <br/>
 The dataset is about the quality of air measured by a gas multisensor device. The data is hourly basis recorded along with the gas concentration references from an analyzer.
The **'Air Quality'** Dataset consists of 9358 instances. The data is averaged from 5 sensors in hourly basis.
The air quality is measured by the number of compounds/elements present in the air.

### Features of Data
<br/>
 The compounds(features) that effect the air quality for that particular data and time is as follows:<br/>
 	
    1. Concentration CO in mg/m^3
	2. PT08.S1 (tin oxide)
	3. Non Metanic HydroCarbons concentration in microg/m^3 
	4. Benzene concentration in microg/m^3
	5. PT08.S2 (titania) 
	6. NOx concentration in ppb 
	7. PT08.S3 (tungsten oxide)
	8. NO2 concentration in microg/m^3 
	9. PT08.S4 (tungsten oxide) 
	10. PT08.S5 (indium oxide) 
	11. Temperature in °C 
	12. Relative Humidity (%) 
	13. Absolute Humidity

### Source of Data
<br/>
The Source of **Air Quality** -- https://archive.ics.uci.edu/ml/datasets/Air+Quality

### Reading of the Data
<br/>
The dataset is in .xlsx format. So I used **'.read_excel'** method to read the data from **'datasetregression/AirQualityUCI.xlsx'** file.

### Preprocessing of Data

### Reading the data
<br/>
The dataset is in .csv format. So I used **'.read_csv'** method to read the data from **'datasetregression/adult.csv'** file.

### Preprocessing of Data
<br />

The dataset doesnot contain headers and row_indexes. So we need to **manually enter all the names of the attributes** and use the option **index_col=False to not treat the first column as row index**. 
<br/><br/>
The columns **'workclass'** and **'Native-Country'** has missing values indicated as **?**. So, we are eliminating such rows.
<br/><br/>
*So now we can plot the data and visualize them.*

```python
names=('age','workclass','fnlwgt','education','education-num','marital status','occupation','relationship','race','sex','capital gain','capital-loss','hrswk','Native-Country','income')

cla_data = pd.read_csv('datasetclassification/adult.csv',names=names,index_col=False)
cla_data = cla_data[~cla_data['workclass'].str.contains('\?', na=False)]
cla_data = cla_data[~cla_data['Native-Country'].str.contains('\?', na=False)]
cla_data
```

### Visualization of Data
<br/><br/>
Below, I have plotted the data for **sex VS income**
<br/>
###### Explaining the work:
1. For sex vs income I have taken the **multi-bar graph** to plot. THe *x-axis* contains the 'SEX' series. The **blue bar** represents the <=50K and **red bar** represents the >50K number.

2. First, I have taken the rows who are 'Male' in *male_data* variable, and divided into data who has less than or equal to 50K category and greater than 50K category.

3. Similary, I divided 'Female' rows too.

4. These were grouped into *means_less_50_income*(male,female) and *means_greater_50_income*(male,female)

5. The data was passed into the **plt.bar** method and plt.show method gets called by jupyter itself because of **(%matplotlib inline)** line.

```python
#sex vs income

#one way to minimize the computing is by taking numbers.
# data to plot
male_data = cla_data[cla_data['sex'].str.contains('Male', na=False)]
male_less_50_data = male_data[male_data['income'].str.contains('<=50K',na=False)]
male_greater_50_data = male_data[male_data['income'].str.contains('>50K',na=False)]

female_data = cla_data[cla_data['sex'].str.contains('Female',na=False)]
female_less_50_data = female_data[female_data['income'].str.contains('<=50K',na=False)]
female_greater_50_data = female_data[female_data['income'].str.contains('>50K',na=False)]

n_groups = 2
means_less_50_income = (male_less_50_data['sex'].count(), female_less_50_data['sex'].count())
means_greater_50_income = (male_greater_50_data['sex'].count(), female_greater_50_data['sex'].count())
 
# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.15
opacity = 0.8
 
rects1 = plt.bar(index, means_less_50_income, bar_width,
                 alpha=opacity,
                 color='b',
                 label='<=50K')
 
rects2 = plt.bar(index + bar_width, means_greater_50_income, bar_width,
                 alpha=opacity,
                 color='g',
                 label='>50K')
 
plt.xlabel('Sex')
plt.ylabel('Income')
plt.title('Sex vs Income')
plt.xticks(index + bar_width, ('Male','Female'))
plt.legend()
```

### Observation


From observing the above graph, itseems like the **male population is very large**. The males are earning more than 4 times than the females.
For the females to have more income, we can develop strategies like giving maternity leaves, decreasing the number of working hours etc., such that they come forward for taking a job. Of course, maintaining equal proportion is good for the country's economy.

### Visualization of Data
<br/><br/>
Below, I have plotted the data for **age VS hrswk**
<br/>
###### Explaining the work:
1. For age vs hrswrk I have taken the **dots graph** to plot. THe *x-axis* contains the 'Age' series and *y-axis* contains the 'Hrswk' series. The **red dots** represents the <=50K and **blue cross** represents the >50K number.

2. First, I have categorized the data into less_50_data variable(The rows contains income <=50K per year) and greater_50_data variable(The rows contains income >50K per year).

3. The data was passed into the **plt.plot** method.

```python
#age vs hrs/wk (based on income classification)
from matplotlib.pyplot import figure
label = ['<=50K','>50k']

less_50_data = cla_data[cla_data['income'].str.contains('<=50K',na=False)]
greater_50_data = cla_data[cla_data['income'].str.contains('>50K',na=False)]
figure(figsize = (15,10))
plt.plot(less_50_data['age'],less_50_data['hrswk'], 'or',label='<=50K')
plt.plot(greater_50_data['age'],greater_50_data['hrswk'],'xb',label = '>50K')

plt.xlabel('Age')
plt.ylabel('Hrswk') 

plt.title('Age vs Hrswk')
plt.legend()
plt.show()
```

### Observation
1. From observing the above graph, **most of the from Age 35-50 years work from 37-55 hrs a week and earn >50K income.**
2. There are few cases where the people are working for more than 90hrs/wk and still earning <=50K
3. Some exceptional people are working for roughly 10hrs/wk and earning >50K per year.
4. One more observation we can make is that, **majority people are working in the early years and slowly that rate (density) decreases as the age increases.**
